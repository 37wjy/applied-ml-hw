```python
import numpy as np
import matplotlib.pyplot as plt


def Read():
    x=[0]
    y=[0]
    i=0
    for line2 in open("data2.txt"):
        #print(line2)
        li=line2.split(',')
        x.append(li[0])
        y.append(li[1][:-1])
    return x,y


def main():
    x,y=Read()
    plt.style.use('seaborn-whitegrid')
    plt.title("buibui")
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.plot(x,y,'bo')
    plt.show()
```
